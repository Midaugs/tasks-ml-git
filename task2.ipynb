```python
import cv2
import numpy as np
import pandas as pd

# Calibration: manually measured from the image
x_pixel_min, x_pixel_max = 100, 600  # X axis from 0 to 500
y_pixel_min, y_pixel_max = 400, 100  # Y axis from 0 to 8

def pixel_to_data(x_px, y_px):
    x_data = (x_px - x_pixel_min) / (x_pixel_max - x_pixel_min) * 500
    y_data = (y_pixel_min - y_px) / (y_pixel_min - y_pixel_max) * 8
    return x_data, y_data

# Load image and process
img = cv2.imread('Y_scatter_size.jpg', cv2.IMREAD_GRAYSCALE)
_, thresh = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY_INV)
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Extract data points
data_points = []
for cnt in contours:
    M = cv2.moments(cnt)
    if M["m00"] != 0:
        cx = int(M["m10"] / M["m00"])
        cy = int(M["m01"] / M["m00"])
        x_val, y_val = pixel_to_data(cx, cy)
        data_points.append((x_val, y_val))

# Save to CSV
df = pd.DataFrame(data_points, columns=["x", "y"])
df = df.sort_values(by="x")  # Optional: sort by X
df.to_csv("Y_scatter_data_converted_size.csv", index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_system(x: np.ndarray, z0=0.0, y0=0.0, y1=0.0):
    """
    Simulate the system with:
    y[i] = y[i-1] + 0.01 * y[i-2] + 8 * x[i-1] - 0.3 * x[i-2] + 0.1 * z[i-1]
    z[i] = z[i-1] + 2 * x[i-1] + 0.11

    Parameters:
        x (np.ndarray): Input signal array
        z0 (float): Initial value of z[-1]
        y0 (float): Initial value of y[-1]
        y1 (float): Initial value of y[-2]

    Returns:
        Tuple[np.ndarray, np.ndarray]: Output arrays for y and z
    """
    n = len(x)
    y = np.zeros(n)
    z = np.zeros(n)
    
    # Initial conditions
    y[0] = y0
    y[1] = y1
    z[0] = z0

    for i in range(2, n):
        z[i] = z[i-1] + 2 * x[i-1] + 0.11
        y[i] = y[i-1] + 0.01 * y[i-2] + 8 * x[i-1] - 0.3 * x[i-2] + 0.1 * z[i-1]

    return y, z
```

```python
import numpy as np

def generate_data(n_steps, target_y):
    x_vals = [0, 0]  # x(i-2), x(i-1)
    y_vals = [0, 0]  # y(i-2), y(i-1)
    z_vals = [0]     # z(i-1)

    for i in range(2, n_steps):
        y_prev_1 = y_vals[-1]
        y_prev_2 = y_vals[-2]
        z_prev = z_vals[-1]
        x_prev_2 = x_vals[-2]

        # Sprendžiame x(i-1), kad y būtų lygus target_y
        x_i_1 = (target_y - y_prev_1 - 0.01 * y_prev_2 - 0.1 * z_prev + 0.3 * x_prev_2) / 8
        x_vals.append(x_i_1)

        y_i = y_prev_1 + 0.01 * y_prev_2 + 8 * x_i_1 - 0.3 * x_prev_2 + 0.1 * z_prev
        y_vals.append(y_i)

        z_i = z_prev + 2 * x_i_1 + 0.11
        z_vals.append(z_i)

    return np.array(x_vals[2:]), np.array(y_vals[2:]), np.array(z_vals[1:])

# Parametrai
n_steps = 100
target_y = 5.0

# Vykdymas
a = generate_data(n_steps, target_y)
print("x:", a[0])
print("y:", a[1])
print("z:", a[2])
```

```python
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split

# Sukuriam tinklą
class InverseModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(4, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.net(x)

# Sugeneruojam duomenis
x, y, z = generate_data(1000, target_y=5)
features = []
targets = []

for i in range(2, len(x)):
    xi_2 = x[i-2]
    yi_1 = y[i-1]
    yi_2 = y[i-2]
    zi_1 = z[i-1]
    features.append([yi_1, yi_2, xi_2, zi_1])
    targets.append(x[i])

X = torch.tensor(features, dtype=torch.float32)
y_target = torch.tensor(targets, dtype=torch.float32).view(-1, 1)

# Apmokymas
model = InverseModel()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
loss_fn = nn.MSELoss()

for epoch in range(1000):
    optimizer.zero_grad()
    y_pred = model(X)
    loss = loss_fn(y_pred, y_target)
    loss.backward()
    optimizer.step()
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: loss = {loss.item():.6f}")
```

```python
# Išsaugoti modelį
torch.save(model.state_dict(), "inverse_model.pth")

# Uzkrovimui
loaded_model = InverseModel()
loaded_model.load_state_dict(torch.load("inverse_model.pth"))
loaded_model.eval()
```

```python
def test_model(loaded_model, steps=300):
    y_vals = [0, 0]
    z_vals = [0]
    x_vals = []

    for i in range(2, steps):
        target_y = 5 if i < 200 else 7
        yi_1 = y_vals[-1]
        yi_2 = y_vals[-2]
        zi_1 = z_vals[-1]
        xi_2 = x_vals[-2] if i >= 4 else 0

        input_tensor = torch.tensor([[yi_1, yi_2, xi_2, zi_1]], dtype=torch.float32)
        xi_1 = loaded_model(input_tensor).item()

        x_vals.append(xi_1)
        y_i = yi_1 + 0.01 * yi_2 + 8 * xi_1 - 0.3 * xi_2 + 0.1 * zi_1
        z_i = zi_1 + 2 * xi_1 + 0.11

        y_vals.append(y_i)
        z_vals.append(z_i)

    return x_vals, y_vals[2:], z_vals[1:]
```

```python
import matplotlib.pyplot as plt

x_out, y_out, z_out = test_model(loaded_model)

plt.figure(figsize=(12, 6))
plt.plot(y_out, label="y(t)")
plt.plot(z_out, label="z(t)")
plt.axvline(200, color='r', linestyle='--', label='Change to y=7')
plt.title("System Output with NN-Controlled x")
plt.xlabel("Time step")
plt.legend()
plt.grid(True)
plt.show()
```
